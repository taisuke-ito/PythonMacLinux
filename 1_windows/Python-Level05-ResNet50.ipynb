#### ■ Tensorflowライブラリのインポート

```python
import tensorflow
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

#### ■ ニューラルネットワークモデルの読み込み
#####         (ResNet50モデル)

```python
model = ResNet50(weights='imagenet')
```

```python
#model.summary()
```

#### ■ 画像の表示
#### IPythonパッケージは、JupyterLabと一緒にインストールされ、Jupyterに画像を表示できます。

###### ↓画像ファイルのパスの文字列を作成

```python
import os
folderA = os.getcwd().replace('\\','/')
folderA
```

```python
file_path = folderA + '/picture/image07.jpg'
print(file_path)
```

###### ↓IPythonパッケージを使って画像の表示

```python
from IPython.display import Image
Image(file_path, width = 200, height = 100)
```

#### ■ 画像データの読み込み

###### kerasを使って、画像データを読み込みます。

```python
img = image.load_img(file_path, target_size=(224, 224))
```

```python
type(img)
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

#### ■ 推論

```python
preds = model.predict(x)
```

```python
#preds
#print(type(preds))
#print(preds.shape)
```

#### ■ 結果の表示

```python
decode_predictions(preds, top=1)[0]
```
